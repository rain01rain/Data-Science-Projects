# Data Cleaning

```python
b['Gender']=b['Gender'].replace(to_replace=np.nan,value='no data')
```

```python
b['Dependents']=b['Dependents'].fillna(b['Dependents'].mode()[0])
```

```python
b['Self_Employed']=b['Self_Employed'].replace(to_replace=np.nan,value='no data')
```

```python
b['LoanAmount']=b['LoanAmount'].fillna(b['LoanAmount'].mean())
```

```python
b['Loan_Amount_Term']=b['Loan_Amount_Term'].fillna(b['Loan_Amount_Term'].mean())
```

```python
b['Credit_History']=b['Credit_History'].fillna(b['Credit_History'].mode()[0])
```

```python
b.isnull().sum()
```

# Data Analysis

```python
sns.relplot(x='ApplicantIncome',y='LoanAmount',data=b)
```

Conclusion: High income applicant will have more chances of getting high loan amount.

```python
sns.pairplot(b)
```

Conclusion: Relationship between different components which shows the chances of getting loan.

```python
sns.relplot(x='ApplicantIncome',y='LoanAmount',data=b,hue='Credit_History')
```

Conclusion: Applicants having credit history 1 have higher chances of getting the loan approval.

```python
sns.relplot(x='ApplicantIncome',y='LoanAmount',data=b,hue='Property_Area')
```

Conclusion: Applicants from Urban area have higher chances of getting the loan approval.

```python
sns.relplot(x='ApplicantIncome',y='LoanAmount',data=b,hue='Gender')
```

Conclusion: Male applicants have more chances of getting loan approval compared to female applicants.

```python
sns.relplot(x='ApplicantIncome',y='LoanAmount',data=b,hue='Married')
```

Conclusion: Married applicants have more chances of getting loan approval compared to unmarried applicants.

```python
sns.relplot(x='ApplicantIncome',y='LoanAmount',data=b,hue='Education')
```

Conclusion: Graduate applicants will have more chances of getting loan approval compared to not graduate applicants. 

```python
sns.relplot(x='ApplicantIncome',y='LoanAmount',data=b,hue='Dependents')
```

Conclusion: Applicants having less number of dependents have higher chances of getting loan approval. 

```python
sns.relplot(x='ApplicantIncome',y='LoanAmount',data=b,hue='Loan_Amount_Term')
```

Conclusion: Applicants having high income have higher chances of getting the loan approval and so they can repay the loan in short loan amount term.  

```python
sns.catplot(x='Credit_History',y='LoanAmount',data=b)
```
